## Tesla events in 2018

```python
tesla_events = ['Jun 29,2010',
 'Jun 22,2012',
 'May 8,2013',
 'May 4,2016',
 'Jun 21,2016',
 'Jul 28,2017',
 'Nov 1,2017',
 'Jan 3,2018',
 'Apr 3,2018',
 'May 2,2018',
 'Jul 1,2018',
 'Aug 1,2018',
 'Aug 7,2018'
]
tesla_event_dates = [datetime.strptime(i,  '%b %d,%Y') for i in tesla_events]
```

```python
def days_after_tesla_event(date_obj):
    dates_diffs = [date_obj-dates for dates in tesla_event_dates]
    return min([days.days for days in dates_diffs if days.days>=0])
```

```python
data['DATES'] = data['DISBURSAL_DATE'].apply(days_after_tesla_event)
```

```python
data['DATES'] = scale_continous(data['DATES'], 'DATES')
data['APPLICANTS_AGE'] = scale_continous(data['APPLICANTS_AGE'], 'APPLICANTS_AGE')
data['DISBURSAL_AGE'] = scale_continous(data['DISBURSAL_AGE'], 'DISBURSAL_AGE')
```
